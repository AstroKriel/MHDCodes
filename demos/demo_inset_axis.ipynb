```python
import numpy as np
import matplotlib.pyplot as plt

from numpy.random import rand
from matplotlib.gridspec import GridSpec

## initialise figure
fig_scales = plt.figure(figsize=(12, 5))
fig_grids = GridSpec(ncols=2, nrows=1, figure=fig_scales)
ax0 = fig_scales.add_subplot(fig_grids[0])
ax1 = fig_scales.add_subplot(fig_grids[1])
## plot data
num_points = 100
x = np.linspace(0, 4*np.pi, num_points) + 2*rand(num_points)
ax0.plot(np.cos(x), "k.")
ax1.plot(np.sin(x), "k.")
## create inset axis
ax0_inset = ax0.inset_axes([0.5, 0.45, 0.5, 0.5])
ax1_inset = ax1.inset_axes([0.0, 1.0, 1.0, 0.5])
ax1_inset.xaxis.tick_top()
## show plot
plt.show()
```
